# Importaciones

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import tensorflow as tf  
from tensorflow import keras
import matplotlib.pyplot as plt 
```

# Modelo

```python
datosRecolectados = pd.read_excel('Datos.xlsx')
X = datosRecolectados[['Peso', 'Tiempo']]
Y = datosRecolectados['Kilocalorias']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=25)
capa_entrada = tf.keras.layers.Dense(units = 10, input_shape = (X.shape[1],))
capa_oculta = tf.keras.layers.Dense(units = 20)
capa_salida = tf.keras.layers.Dense(units = 1)
modelo = tf.keras.Sequential([capa_entrada, capa_oculta,capa_salida])

modelo.compile(optimizer = 'adam', loss = 'mean_squared_error')

historial = modelo.fit(X_train, y_train, epochs = 1100, batch_size = 1)
```

# Conjunto de Pruebas

```python
# Caso 1
datosPrueba_1 = [[75, 30]]
resultado_1 = modelo.predict(datosPrueba_1)
print("Una persona con un peso de", datosPrueba_1[0][0], "kg y un tiempo de", datosPrueba_1[0][1], "minutos corriendo a 9.6 km/h")
print("Sus Kilocalorias son: ", resultado_1)
```

```python
# Caso 2
datosPrueba_2 = [[50, 45]]
resultado_2 = modelo.predict(datosPrueba_2)
print("Una persona con un peso de", datosPrueba_2[0][0], "kg y un tiempo de", datosPrueba_2[0][1], "minutos corriendo a 9.6 km/h")
print("Sus Kilocalorias son: ", resultado_2)
```

```python
# Caso 3
datosPrueba_3 = [[90, 60]]
resultado_3 = modelo.predict(datosPrueba_3)
print("Una persona con un peso de", datosPrueba_3[0][0], "kg y un tiempo de", datosPrueba_3[0][1], "minutos corriendo a 9.6 km/h")
print("Sus Kilocalorias son: ", resultado_3)
```

# Grafica

```python
# Evaluacion del modelo
perdida = modelo.evaluate(X_test, y_test)
print("La funcion de perdida es: ", perdida)
plt.xlabel("Vueltas")
plt.ylabel("Rango de perdida")
plt.plot(historial.history["loss"])
```
